```python
# Convert the 'Scheduled Date' column to datetime format for comparison
# 'Not Scheduled' or any invalid date will be converted to NaT
df['Scheduled Date'] = pd.to_datetime(df['Scheduled Date'], format='%d/%m/%Y', errors='coerce')

# Define a mapping of years to their respective columns
year_to_column_map = {
    2014: '2014',
    2015: '2015',
    2016: '2016',
    2017: '2017',
    2018: '2018',
    2019: '2019',
    2020: '2020 F2D',
    2021: '2021 F2D',
    2022: '2022 F2D',
    2023: '2023 F2D',
    2024: '2024 F2D'
}

# Create a function to update the master planner with workshop dates
def update_planner(planner_df, workshops_df):
    for _, workshop in workshops_df.iterrows():
        # Extract school name and date
        school_name = workshop['School Name']
        workshop_date = workshop['Scheduled Date']

        # If the workshop date is NaT (invalid), skip this entry
        if pd.isna(workshop_date):
            continue

        # Extract the year from the date
        year = workshop_date.year

        # Find the matching row in the planner DataFrame
        if school_name in planner_df['Organisation Name'].values:
            row_index = planner_df[planner_df['Organisation Name'] == school_name].index[0]
            
            # Map the year to the correct column
            if year in year_to_column_map:
                column_name = year_to_column_map[year]
                # Update the relevant year column with the workshop date
                planner_df.at[row_index, column_name] = workshop_date.strftime('%d/%m/%Y')

    return planner_df
```

```python
# Update the planner DataFrame with workshop dates
df_updated_planner = update_planner(df_schools, df)

# Save the updated planner DataFrame to a new CSV file
df_updated_planner.to_csv('updated_master_planner.csv', index=False)

print("Master planner updated and saved to 'updated_master_planner.csv'")
```
